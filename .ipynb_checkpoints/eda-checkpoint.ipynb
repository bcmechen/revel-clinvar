## 2. Exploratory data analysis.

**Distribution of REVEL scores for variants in ACMG59 genes.**

```python
plt.figure(figsize=(16, 6))
sns.histplot(data=df, x='REVEL', stat='probability', hue='CLASS', common_norm=False, multiple='dodge', 
             hue_order=['Benign', 'VUS', 'Pathogenic', 'Other'],
             palette=['Green', 'White', 'Maroon', 'Gray'], bins=20)
plt.xlabel('REVEL Score', fontsize=16)
plt.ylabel('Proportion', fontsize=16)
plt.tick_params(axis='both', labelsize=12)
plt.legend(labels=['Other', 'Pathogenic', 'VUS', 'Benign'], frameon=False, fontsize=14)
```

**Sensitivity and specificity for different REVEL score thresholds.**

```python
# Subsetting pathogenic and benign data.
two_class_df = df[(df['CLASS'] == 'Benign') | (df['CLASS'] == 'Pathogenic')]

# Calculate the True Positive Rates, False Positive Rates and Thresholds.
fpr, tpr, thresholds = roc_curve(two_class_df['CLASS'], two_class_df['REVEL'], pos_label='Pathogenic')

plt.figure(figsize=(8, 6))
plt.plot(thresholds[1:], tpr[1:], color='red')
plt.plot(thresholds[1:], 1-fpr[1:], color='blue')
plt.xlabel('REVEL Score Threshold', fontsize=16)
plt.ylabel('Sensitivity and Specificity', fontsize=16)
plt.tick_params(axis='both', labelsize=12)
plt.legend(labels=['Sensitivity', 'Specificity'], frameon=False, fontsize=14)
```

**Performance of REVEL in loss-of-function (lof) genes and non-lof genes.**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Subsetting genes with haploinsufficiency scores that are 3, 2 (evidence for dosage pathogenicity).
sns.boxplot(data=df[df['Haploinsufficiency Score'].isin([3, 2])], x='CLASS', y='REVEL', ax=ax1, 
            width=0.4, fliersize=0, order=['Benign', 'VUS', 'Pathogenic', 'Other'], 
            palette=['Green', 'White', 'Maroon', 'Gray'])
ax1.set_xlabel('')
ax1.set_ylabel('REVEL Score', fontsize=14)
ax1.tick_params(labelsize=14)
ax1.set_title('Haploinsufficient genes (N=48369)', fontsize=16)

# Subsetting genes with haploinsufficiency scores that are 0, 40 (no evidence for dosage pathogenicity).
sns.boxplot(data=df[df['Haploinsufficiency Score'].isin([40, 0])], x='CLASS', y='REVEL', ax=ax2, 
            width=0.4, fliersize=0, order=['Benign', 'VUS', 'Pathogenic',  'Other'], 
            palette=['Green', 'White', 'Maroon', 'Gray'])
ax2.set_xlabel('')
ax2.set_ylabel('REVEL Score', fontsize=14)
ax2.tick_params(labelsize=14)
ax2.set_title('Non-haploinsufficient genes (N=8409)', fontsize=16)
```

**Performance of REVEL in individual ACMG59 gene.**

Calculate AUROC and AUPRC for each gene.

```python
revel_performance_by_gene = {'gene': [], 'AUROC': [], 'AUPRC': []}
for gene in df['GENE'].unique():
    # Use two_class subset.
    # Further subsetting by gene.
    gene_df = two_class_df[two_class_df['GENE'] == gene]
    try:
        p_count = gene_df['CLASS'].value_counts()['Pathogenic']
    except KeyError:
        p_count = 0
    try:
        b_count = gene_df['CLASS'].value_counts()['Benign']
    except KeyError:
        b_count = 0
    revel_performance_by_gene['gene'].append('{} ({} P; {} B)'.format(gene, p_count, b_count))
    # Calculate TPR and FPR.
    fpr, tpr, thresholds = roc_curve(gene_df['CLASS'], gene_df['REVEL'], pos_label='Pathogenic')
    revel_performance_by_gene['AUROC'].append(auc(fpr, tpr))
    # Calculate recall and precision.
    precision, recall, thresholds = precision_recall_curve(gene_df['CLASS'], gene_df['REVEL'], pos_label='Pathogenic')
    revel_performance_by_gene['AUPRC'].append(auc(recall, precision))
```

Plot AUROC and AUPRC for each gene.

```python
gene_performance_df = pd.DataFrame(revel_performance_by_gene)
# Drop ACTA2 gene because there is no benign variant in ACTA2.
gene_performance_df = gene_performance_df[~gene_performance_df['gene'].str.contains('ACTA2')]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, 16))

sns.barplot(x='AUROC', y='gene', data=gene_performance_df.sort_values(by=['AUROC'], ascending=False), 
            orient='h', ax=ax1, color='c')
ax1.set_xlabel('AUROC', fontsize=20)
ax1.set_ylabel('')
ax1.tick_params(axis='x', labelsize=18)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

sns.barplot(x='AUPRC', y='gene', data=gene_performance_df.sort_values(by=['AUPRC'], ascending=False), 
            orient='h', ax=ax2, color='c')
ax2.set_xlabel('AUPRC', fontsize=20)
ax2.set_ylabel('')
ax2.tick_params(axis='x', labelsize=18)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
```
